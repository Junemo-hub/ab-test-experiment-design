## 1. Experiment Overview

- **Objective:** To evaluate whether the new landing page leads to a higher conversion rate compared to the old one.
- **Group Allocation:**
  Control: Old landing page
  Treatment: New landing page

---

## 2. üìä Summary of Experimental Data

| Group      | Conversions | Total Users | Conversion Rate (%) |
|------------|-------------|-------------|----------------------|
| Control    | 17,489      | 145,274     | 12.04%               |
| Treatment  | 17,264      | 145,315     | 11.89%               |

> Dataset: [Kaggle - Ecommerce A/B Testing (2022)](https://www.kaggle.com/datasets/putdejudomthai/ecommerce-ab-testing-2022-dataset1/data)

## 3. üî¨ Statistical Analysis

- **Z-Test (Proportion Test):**
  - Z-statistic: `1.3143`
  - P-value: `0.1887`
  - ‚ùå No statistically significant difference at Œ± = 0.05

- **Chi-Square Test:**
  - œá¬≤-statistic: `1.7125`
  - P-value: `0.1907`

- **Effect Size (Cohen's h):** `0.0049`
  - üü° Very small effect (less than 0.2)              

- **95% Confidence Intervals for Conversion Rates:**
  - Control: [0.1187, 0.1221]
  - Treatment: [0.1171, 0.1205]  
  ‚Üí Overlapping confidence intervals indicate **no significant difference** in conversion rates between the groups.

## 5. Conclusion & Insights

- ‚úÖ The control group showed a slightly higher conversion rate (12.04%) than the treatment group (11.89%).
- ‚ùå However, since the **p-value is 0.1887**, there is **no statistically significant difference** between the two groups.
- üéØ **Conclusion:** The new landing page does not show a clear performance improvement over the old one.  
‚Üí It is recommended to **continue using the current page for now**.
