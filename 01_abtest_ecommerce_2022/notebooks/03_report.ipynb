# 📘 Final Report: A/B Test Analysis

---

## 1. 🎯 Experiment Overview

- **Objective:** To evaluate whether the new landing page leads to a higher conversion rate compared to the old one.
- **Group Allocation:**
- **Control:** Old landing page
- **Treatment:** New landing page

---

## 2. 📊 Summary of Experimental Data

| Group      | Conversions | Total Users | Conversion Rate (%) |
|------------|-------------|-------------|----------------------|
| Control    | 17,723      | 147,202     | 12.04%               |
| Treatment  | 17,514      | 147,280     | 11.89%               |

> Dataset: Kaggle - Ecommerce A/B Testing (2022)

---

## 3. 🔬 Statistical Analysis

- **Z-Test (Proportion Test):**
  - Z-statistic: `1.2396`
  - P-value: `0.2151`
  - ❌ No statistically significant difference at α = 0.05

- **Chi-Square Test:**
  - χ²-statistic: `1.5227`
  - P-value: `0.2172`

- **Effect Size (Cohen's h):** `0.0046`
  - 🟡 Very small effect (less than 0.2)

- **95% Confidence Intervals for Conversion Rates:**
  - Control: [0.1187, 0.1221]
  - Treatment: [0.1173, 0.1206]  
  → Overlapping confidence intervals indicate **no significant difference** in conversion rates between the groups.

## 5. 🧠 Conclusion & Insights

- ✅ The control group showed a slightly higher conversion rate (12.04%) than the treatment group (11.89%).
- ❌ However, since the **p-value is 0.2151**, there is **no statistically significant difference** between the two groups.
- 🎯 **Conclusion:** The new landing page does not show a clear performance improvement over the old one.  
→ It is recommended to **continue using the current page for now**.

- Run follow-up experiments including additional KPIs such as time on page or scroll depth.
- Consider subgroup A/B testing by country or device type to detect regional effects.
