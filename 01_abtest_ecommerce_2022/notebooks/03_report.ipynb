# 📘 A/B 테스트 최종 보고서

---

## 1. 🎯 실험 개요

- **목적:** 새 landing page가 기존보다 더 높은 전환율을 유도하는지 검증  
- **실험군 분할:**
  - Control: 기존 페이지
  - Treatment: 새로운 페이지

---

## 2. 📊 실험 데이터 요약

| 그룹       | 전환 수 | 전체 수 | 전환율 (%) |
|------------|---------|---------|------------|
| Control    | 17,723  | 147,202 | 12.04%     |
| Treatment  | 17,514  | 147,280 | 11.89%     |

> 데이터셋: Kaggle - Ecommerce AB Test (2022)

---

## 3. 🔬 통계 분석 결과

- **Z-Test 결과:**
  - Z 통계량: `1.2396`
  - P-value: `0.2151`
  - ❌ 유의수준 0.05 기준 **차이 없음**

- **Chi-Square Test:**
  - χ² 통계량: `1.5227`
  - P-value: `0.2172`

- **Cohen's h (효과 크기):** `0.0046`
  - 🟡 아주 미미한 효과 (0.2 미만)

## 5. 🧠 결론 및 인사이트

- ✅ Control 그룹의 전환율(12.04%)이 Treatment(11.89%)보다 소폭 높았음
- ❌ 그러나 **p-value가 0.2151로 유의수준 0.05보다 크므로** 통계적으로 유의미한 차이는 없음
- 🎯 결론: **새로운 페이지가 기존 대비 성능 향상이 확실하지 않음**  
→ 당장은 기존 페이지 유지 추천

---

## 6. 📌 제안 사항

- 체류 시간 등 다른 KPI를 포함한 후속 실험 필요
- 국가별로 사용자 반응 차이가 있는지 subgroup A/B 테스트 진행 제안
- UI/UX 요소 외에 다른 전환 유도 요소(추천 상품, CTA 문구 등) 테스트 권장
